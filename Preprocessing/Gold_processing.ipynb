```python
import matplotlib.pyplot as plt
lags = [10,20]
# Ãp dá»¥ng Ä‘á»™ trá»… (lag) cho giÃ¡ vÃ ng
df_lagged = df_AHT[['Close']].copy()
for lag in lags:
    df_lagged[f'Gold_Lag{lag}'] = df_gold['EUR'].shift(lag)

# Váº½ biá»ƒu Ä‘á»“
plt.figure(figsize=(12, 6))
plt.plot(df_lagged.index, df_lagged['Close'], label='AHT Close', linewidth=2)

for lag in lags:
    plt.plot(df_lagged.index, df_lagged[f'Gold_Lag{lag}'], label=f'Gold Lag {lag}', linestyle='--')

plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Gold Price (Lagged) vs AHT Close Price')
plt.legend()
plt.grid(True)
plt.show()
```
